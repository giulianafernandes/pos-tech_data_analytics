# **AULA 3** - Manipulação de Dados

```python
# para selecionar duas colunas
dados[['2008/Ago', '2008/Set']].head()
```

```python
dados.head()
```

```python
# mostra o nome de todas as colunas
dados.columns
```

```python
# essa função só retorna dados de colunas em que existem dados inseridos, quando a coluna não tem valores não existe média
dados.mean()
```

```python
# portanto abaixo tiramos a média novamente e geramos uma lista de python que nos devolve todas as colunas com valores existentes
colunas_usaveis = dados.mean().index.tolist()

# agora além das colunas usáveis vamos inserir os nomes dos estados
colunas_usaveis.insert(0, "Unidade da Federação")

colunas_usaveis[:5]
```

```python
# atribuímos um novo nome ao framework agora com as colunas sem espaços em branco
dados_usaveis = dados[colunas_usaveis]
dados_usaveis.head()
```

```python
# para tornar cada unidade da federação em um índice utilizamos set_index (dessa forma substituímos os números a esquerda pelo nome da unidade da federação)
dados_usaveis = dados_usaveis.set_index("Unidade da Federação")
dados_usaveis.head()
```

```python
# quando usamos colchetes estamos fazendo uma busca pelas colunas
dados_usaveis["2019/Ago"].head()
```

```python
# para localizar valores pelo índice usamos .loc[]      - busca nas linhas pelo nome
dados_usaveis.loc["12 Acre"]
```

```python
# para localizar dados pelo numero da linha usamos .iloc[]  - lembrando que a contagem em python começa sempre em 0
dados_usaveis.iloc[1]
```

```python
dados_usaveis.plot(figsize=(12,6))
```

```python
# para transpor dados - linha vira coluna e coluna vira linha
dados_usaveis.T.head()
```

```python
dados_usaveis.T.plot(figsize=(12,6))
```

```python
# o gráfico ficou com uma linha discrepante no final porque o total é muito maior do que todo o restante do gráfico
# neste caso precisamos remover a ultima coluna, então usamos .drop
dados_usaveis = dados_usaveis.drop("Total", axis=1)
dados_usaveis.head()
```

```python
dados_usaveis.T.plot(figsize=(12,6))
```

```python
dados_usaveis.T
```

# **AULA 4** - Análise de Gráficos e Criação de Hipóteses

```python
# formatanto o gráfico
ordenados_por_total = ordenados_por_total / 1_000_000
```

```python
axis = ordenados_por_total.head(5).T.plot(figsize=(10,6))
# axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
axis.set_ylabel("gasto aprovado em milhões de reais")
axis.set_xlabel("ano/mês")
```

```python
#criamos uma variável tirando os 47 primeiros meses da visualização
meses_filtrados = ordenados_por_total.columns[47:]
ordenados_por_total[meses_filtrados].head(3).T.plot(figsize=(10,6))
axis.set_ylabel("gasto aprovado em milhões de reais")
axis.set_xlabel("ano/mês")
```

```python
# definimos uma função para facilitar quando plotarmos os dados novamente
def plota_gastos_por_mes(dados):
  axis = dados.T.plot(figsize=(10,6))
  axis.set_ylabel("gasto aprovado em milhões de reais")
  axis.set_xlabel("ano/mês")
```

Agora manipulamos o gráfico gerando diferentes visualizações, dando diferentes impressões. É importante pensar bem no tipo de mensagem que gostaríamos de transmitir e se essa mensagem é verdadeira ou está enviesada.

```python
plota_gastos_por_mes(ordenados_por_total[meses_filtrados].head(3))
plt.ylim(0,650)
```

```python
plota_gastos_por_mes(ordenados_por_total[ordenados_por_total.columns[45:70]].head(3))
plt.ylim(0,350)
```

```python
plota_gastos_por_mes(ordenados_por_total[ordenados_por_total.columns[-12:]].head(3))
plt.ylim(0,640)
```

```python
plota_gastos_por_mes(ordenados_por_total[ordenados_por_total.columns[-12:]].head(3))
plt.ylim(80,640)
```

```python
plota_gastos_por_mes(ordenados_por_total[ordenados_por_total.columns[-60:]].head(3))
plt.ylim(0,640)
```

```python
def plota_gastos_por_mes(dados, figsize=(10,6)):
  axis = dados.T.plot(figsize=figsize)
  axis.set_ylabel("gasto aprovado em milhões de reais")
  axis.set_xlabel("ano/mês")
```

```python
plota_gastos_por_mes(ordenados_por_total.head(5), figsize=(10,6))
plt.ylim(0,640)
```
